Project- 'Should i buy the cloth'.

In this we are trying to make a machine learning model which will output "YES" or "NO" based on the user given parameters. (yes-buy it!, no-don't buy it). we need to manipulate or label the kaggle dataset "Fashion Product Images (Small). we will here label each cloth as "good", "okay", "bad".

```python
import pandas as pd
```

```python
df = pd.read_csv('styles.csv', on_bad_lines='skip')
df.head()
```

```python
df = df.dropna(subset=['usage','season','baseColour'])
print(df.isna().sum())
```

```python
def assign_label(row):
  if row['usage']=='Casual' and row['season'] in ['Summer','All Year'] and row['baseColour'] in ['Black','Blue','White']:
    return 'good'
  elif row['usage']=='Sports' or row['season']=='Winter':
    return 'okay'
  else:
    return 'bad'
```

```python
df['rating']=df.apply(assign_label,axis=1)
df.head()
```

```python
df['rating'].value_counts()
```

```python
#Save the new CSV
df.to_csv('Labeled_fashion_data.csv', index=False)
```
