Project- 'Should i buy the cloth'.

In this we are trying to make a machine learning model which will output "YES" or "NO" based on the user given parameters. (yes-buy it!, no-don't buy it). we need to manipulate or label the kaggle dataset "Fashion Product Images (Small). we will here label each cloth as "good", "okay", "bad".

since the data has more than one classification (good,okay,bad) we need to use multi-class classification model (i.e,

```python
import pandas as pd
```

```python
df = pd.read_csv("Labeled_fashion_data.csv")
df.head()
```

```python
df.drop("id",axis=1, inplace=True)
df.drop("productDisplayName",axis=1, inplace=True)
df.dropna(subset=["rating"],inplace=True)
df.head()
```

```python
print(df["rating"].isna().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder={}
for i in df.columns:
    if df[i].dtype=="object":
        le = LabelEncoder()
        df[i] = le.fit_transform(df[i])
        label_encoder[i] = le
```

```python
df.head()
```

```python
x=df.drop("rating",axis=1)
y=df["rating"]
```

```python
#cross_validation_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=500)
print(cross_val_score(model,x,y,cv=5).mean())
```

```python
model.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x)
print(accuracy_score(y,y_pred))
```

The model has trained now and it got 80.9% accuracy.
Now we proceed to saving the model using pickel

```python
#save the model
import pickle
from pickle import dump

with open("LabelEncoder.pkl","wb") as file:
  pickle.dump(label_encoder,file)

with open("SIBTC_model.pkl","wb") as file:
  pickle.dump(model,file)
```
